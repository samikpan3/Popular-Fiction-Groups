```python
#Importing required libraries
import pandas as pd
import numpy as np
import altair as alt
```

```python
#Getting the dataset into pandas
location = '/Users/samikshyapandey/Downloads/'
document1 = 'df_booksfinalfic.csv'
books = location + document1
df_books = pd.read_csv(books, low_memory=False)
df_books.head()
```

```python
#what I am hoping to do: have drop down for season selection and button for major genre selection.
#maybe a histogram for check out numbers but lets see
```

```python
'''
To use my drop down effectively, i need to filter my data more. So lets remove all other genres expect the ones 
that I need. From previous analysis, we know that the most popular genre were "Mystery, Romance, Suspense and Science Fiction"
So lets filter that dataframe to include only rows that have value
'''
```

```python
most_pop = ['Mystery','Romance','Suspense','Science Fiction']
filtered_df = df_books[df_books['Major_Genre_1'].isin(most_pop)]
filtered_df.head()
```

```python
filtered_df.rename(columns = {"Major_Genre_1": "Genre"}, inplace = True)
```

```python
filtered_df.head()
```

```python
most_pop = ['Mystery','Romance','Suspense','Science Fiction']
season_val = df_books['Season'].unique()
# single-value selection over [Season and Major Genre 1 ] pairs
selection = alt.selection_point(
    name='Select',
    fields=['Season', 'Genre'],
    bind={'Season': alt.binding_select(options=season_val), 'Genre': alt.binding_radio(options=most_pop)}
)
  
# scatter plot, modify opacity based on selection
main_chart = alt.Chart(filtered_df).mark_bar().add_params(
    selection
).encode(
    alt.X('Genre:N', sort=alt.EncodingSortField(field='Checkouts', order='descending')),
    alt.Y('Checkouts:Q', axis=alt.Axis(tickCount=10)),
    tooltip='Title:N',
    opacity=alt.condition(selection, alt.value(0.85), alt.value(0.20)),
    color = alt.Color('Genre:N', legend = None)
).properties(width = 250, height = 350, title = "Most popular genres of Fiction Books checkout at Seattle Public Library")


main_chart
```

```python
main_chart.save('booksborrowfinal.html')
```

```python
brush = alt.selection_interval(
    encodings=['x'] # limit selection to x-axis (year) values
)

years = alt.Chart(filtered_df).mark_bar().add_params(
    brush
).encode(
    alt.X('Month:N', title='Films by Release Year'),
    alt.Y('count(Checkouts):Q', title=None)
).properties(
    width=650,
    height=50
)

years

alt.vconcat(years, main_chart).properties(spacing=5)
```

```python
import altair as alt

# Sample data
most_pop = ['Mystery', 'Romance', 'Suspense', 'Science Fiction']

# Define a custom color scale for genres
genre_colors = alt.Scale(domain=most_pop, range=['#1f77b4', '#ff7f0e', '#2ca02c', '#9467bd'])

# Create a selection
selection = alt.selection_multi(fields=['Major_Genre_1'], bind='legend')

# Create the interactive scatter plot
chart = alt.Chart(filtered_df).mark_point().encode(
    x=alt.X('Checkouts:Q', title='Number of Checkouts'),
    y=alt.Y('Major_Genre_1:N', title='Major Genre'),
    color=alt.Color('Major_Genre_1:N', scale=genre_colors),
    size=alt.Size('Checkouts:Q', title='Number of Checkouts'),
    tooltip=['Title:N', 'Major_Genre_1:N', 'Season:N', 'Checkouts:Q'],
).properties(
    width=600,
    height=300,
    title='Book Checkouts by Major Genre and Season',
).add_selection(selection)

# Add the legend
legend = alt.Chart(filtered_df).mark_point().encode(
    y=alt.Y('Major_Genre_1:N', axis=None),
    color=alt.Color('Major_Genre_1:N', scale=genre_colors),
).properties(
    title='Major Genre',
    height=30,
    width=200,
).add_selection(selection)

# Combine the chart and legend
final_chart = alt.vconcat(chart, legend)

final_chart
```

```python
location = '/Users/samikshyapandey/Downloads/'
document2 = 'df_foodandcost.csv'
cost_afford = location + document2
df_costafford = pd.read_csv(cost_afford)
df_costafford.head()

```

```python
df_costafford["percent_unable"].describe()
```

```python
df_costafford["Continent"].unique()
df_costafford.head()
```

```python
bin_edges = [0, 25, 50, 100]

# Create categories using cut and custom bin edges
df_costafford['Affordability_Category'] = pd.cut(df_costafford['percent_unable'], bins=bin_edges, labels=['Low', 'Moderate', 'High'])
```

```python
continent = ['EU', 'AF', 'SA', 'AS', 'OC']
affordability = ["Low", "Moderate","High"]
# single-value selection over [Season and Major Genre 2 ] pairs
selection = alt.selection_point(
    name='Select',
    fields=['Continent', 'Affordability_Category'],
    bind={'Continent': alt.binding_select(options=continent), 'Affordability_Category': alt.binding_radio(options=affordability)}
)
  
# scatter plot, modify opacity based on selection
alt.Chart(df_costafford).mark_point(clip = True).add_params(
    selection
).encode(
    alt.Y('PCI increase:Q'),
    alt.X('Healthy Diet Cost:Q',scale=alt.Scale(domain=[-100, 500])),
    tooltip='percent_unable:Q',
    opacity=alt.condition(selection, alt.value(0.75), alt.value(0.05)),
    color = alt.Color('Year:N')
).properties(width = 450, height = 150)
```

This graph plots number of checkouts against book genres. I wanted to investigate among popular fiction books genres, which genres were most popular in Seattle Public Library. Therefore, Genre and Checkouts were placed in position axis. I also wanted to know if we could observe any difference in checked out books and genre across season. Previously, I had added this information by encoding season in facet. Now, through dropdown option, users can decide the season themselves.  Because I have preserved total checkout against genres through opacity, users can now easily compare the ratio of book borrowed in that particular season against total books borrowed of that genre in 2022. Furthermore, I also added selection of genres as a radio bottons, that way, if a user wants to know the title of the books that were borrrowed in a particular season across specific genre they can do so.
